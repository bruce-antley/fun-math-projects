```python
import matplotlib.pyplot as plt
steps = 0
list_of_steps = []
list_of_terms = []
starting_term = 9

list_of_terms.append(starting_term)
list_of_steps.append(steps)
if starting_term % 2 == 0:
    next_term = starting_term // 2
else:
    next_term = starting_term * 3 + 1
steps += 1
list_of_terms.append(next_term)
list_of_steps.append(steps)
#print(next_term)
while next_term != 1:
    if next_term % 2 == 0:
        next_term = next_term // 2
    else:
        next_term = next_term * 3 + 1
    steps += 1
    list_of_terms.append(next_term)
    list_of_steps.append(steps)
#print(list_of_steps)
#print(list_of_terms)
print("Number of steps: " + str(steps))
print("Max in range: " + str(max(list_of_terms)))

#plotting
plt.plot(list_of_steps, list_of_terms)
plt.title("Steps and Range")
plt.xlabel("Steps")
plt.ylabel("Range")
plt.show()
```
